```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_read = cv2.imread('photo.jpg')
img_read = cv2.cvtColor(img_read, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_read)
plt.show()
```

```python
img_read = cv2.cvtColor(img_read, cv2.COLOR_RGB2HSV)
h,s,v = cv2.split(img_read)

value_change = 50
limit = 255 - value_change

v[v > limit] = 255
v[v <= limit] += value_change

img_read = cv2.merge((h,s,v))
img_read = cv2.cvtColor(img_read, cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(img_read)
plt.show()
```

```python
h,w,s = img_read.shape
```

```python
center_w = int(w/2)
center_h = int(h/2)
rotation_angle = 45
```

```python
rot_matrix = cv2.getRotationMatrix2D((center_w,center_h), rotation_angle, 1.0)
img_read = cv2.warpAffine(img_read, rot_matrix, (w,h), borderMode=cv2.BORDER_REFLECT)
```

```python
plt.imshow(img_read)
plt.show()
```
